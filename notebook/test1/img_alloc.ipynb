```python
import cv2 
from FaceLandmark import *

from PoseEstimation import *
```

```python
pose = PoseEstimation()
landmark  = FaceLandmark()
```

```python
import os 
  
# Directory 
lists = ["l1","l2","l3","l4","l5","l6","l7","r1","r2","r3","r4","r5","r6","r7","n"]
for i in lists:
    parent_dir = "../../data/test1/{}".format(i)
    path = os.path.join(parent_dir, "shape_normalized_images") 
    os.mkdir(path)
```

```python
video = cv2.VideoCapture("../../data/test1/2.mov")
success = True
    
```

```python
count = 0
while success:
        success,image = video.read()
        if success:
            processed_image = landmark.process(image)
            #Толгойн чиглэлийг тооцох
            face_features = pose.extract_features(image, processed_image)
            normalized = pose.normalize(face_features)
            pitch_pred, yaw_pred, roll_pred = pose.model.predict(normalized).ravel()

            if yaw_pred > 0.1 and yaw_pred <= 0.2:
                cv2.imwrite("../../data/test1/r1/1_frame_{}.jpg".format(count),image)
            elif yaw_pred > 0.2 and yaw_pred <= 0.3:
                cv2.imwrite("../../data/test1/r2/1_frame_{}.jpg".format(count),image)
            elif yaw_pred > 0.3 and yaw_pred <= 0.4:
                cv2.imwrite("../../data/test1/r3/1_frame_{}.jpg".format(count),image)
            elif yaw_pred > 0.4 and yaw_pred <= 0.5:
                cv2.imwrite("../../data/test1/r4/1_frame_{}.jpg".format(count),image)
            elif yaw_pred > 0.5 and yaw_pred <= 0.6:
                cv2.imwrite("../../data/test1/r5/1_frame_{}.jpg".format(count),image)
            elif yaw_pred > 0.6 and yaw_pred <= 0.7:
                cv2.imwrite("../../data/test1/r6/1_frame_{}.jpg".format(count),image)
            elif yaw_pred > 0.7 and yaw_pred <= 0.8:
                cv2.imwrite("../../data/test1/r7/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.1 and yaw_pred >= -0.2:
                cv2.imwrite("../../data/test1/l1/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.2 and yaw_pred >= -0.3:
                cv2.imwrite("../../data/test1/l2/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.3 and yaw_pred >= -0.4:
                cv2.imwrite("../../data/test1/l3/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.4 and yaw_pred >= -0.5:
                cv2.imwrite("../../data/test1/l4/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.5 and yaw_pred >= -0.6:
                cv2.imwrite("../../data/test1/l5/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.6 and yaw_pred >= -0.7:
                cv2.imwrite("../../data/test1/l6/frame_{}.jpg".format(count),image)
            elif yaw_pred < -0.7 and yaw_pred >= -0.8:
                cv2.imwrite("../../data/test1/l7/1_frame_{}.jpg".format(count),image)
            elif yaw_pred < 0.1 and yaw_pred > -0.1:
                cv2.imwrite("../../data/test1/n/1_frame_{}.jpg".format(count),image)
            count += 1
```
